# **Plotting**

```
grouped = df.groupby('province')[['no_grass','no_tiles','no_concrete']].sum()
grouped
```

```
grouped.sort_values(by = 'no_grass', ascending = True)
```

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
grouped.plot(kind = 'bar' ,figsize = (12,12),logy = True)
plt.title('Comparison of roofing materials accross the provinces', color = 'darkblue',fontsize=12)
plt.xlabel('Province',color = 'darkblue',fontsize = 12)
plt.ylabel('number of roofing materials',color='darkblue',fontsize = 12)
plt.legend(['Grass','Titles','Concrete'])
plt.xticks(rotation = 45)
plt.show()
```
